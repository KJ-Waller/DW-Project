```python
areaidx_bbga = bbga.set_index('gebiedcode15')
mask = areaidx_bbga.index.str.match('DX')
```

```python
dxareas_bbga = areaidx_bbga.loc[mask]
```

```python
dxareas_bbga.loc[dxareas_bbga['variabele'].str.match('BEV')]
```

```python
bev_mask = dxareas_bbga.variabele.str.contains('BEV\d?[0|5]_\d?[9|4]{1}\Z')
# incomplete_bev_mask = dxareas_bbga.variabele.str.contains('BEV\d{1}_\d+\Z')

# incomplete_rows = dxareas_bbga.loc[incomplete_bev_mask].map(lambda x: x.variabele = )

DX20_age_groups = dxareas_bbga.loc[(bev_mask) & (dxareas_bbga['jaar'] == 2015) & (dxareas_bbga.index == "DX20")]
DX20_age_groups.reset_index(inplace=True)
drop_idx = DX20_age_groups.loc[(DX20_age_groups['variabele'] == 'BEV65_74') | (DX20_age_groups['variabele'] == 'BEV40_64')].index
DX20_age_groups.drop(axis='index', index=drop_idx, inplace=True)
DX20_age_groups.loc[DX20_age_groups['variabele'] == 'BEV0_4', 'variabele'] = 'BEV00_04'
DX20_age_groups.loc[DX20_age_groups['variabele'] == 'BEV5_9', 'variabele'] = 'BEV05_09'
DX20_age_groups = DX20_age_groups.sort_values('variabele').set_index('variabele')
DX20_age_groups
```

```python
# DX20_age_groups.plot.bar(y='waarde')
DX20_age_groups.plot.barh(y='waarde')
```

```python
def age_dist(region, year):
    bev_mask = dxareas_bbga.variabele.str.contains('BEV\d?[0|5]_\d?[9|4]{1}\Z')

    age_groups = dxareas_bbga.loc[(bev_mask) & (dxareas_bbga['jaar'] == year) & (dxareas_bbga.index == region)]
    age_groups.reset_index(inplace=True)
    drop_idx = age_groups.loc[(age_groups['variabele'] == 'BEV65_74') | (age_groups['variabele'] == 'BEV40_64')].index
    age_groups.drop(axis='index', index=drop_idx, inplace=True)
    age_groups.loc[age_groups['variabele'] == 'BEV0_4', 'variabele'] = 'BEV00_04'
    age_groups.loc[age_groups['variabele'] == 'BEV5_9', 'variabele'] = 'BEV05_09'
    age_groups = age_groups.sort_values('variabele').set_index('variabele')

    age_groups.plot.barh(y='waarde')
```

```python
age_dist('DX19', 2018)
```
